# 数据整合

```python
fields_all = fields1 + fields2 +  fields3 +  fields4 +  fields5 +  fields6 +  fields7 +  fields8 +  fields9
print(fields_all)
df_target = df[fields_all]

df_target.info()
```

**数据处理实现目标**
- 将`iyear``imonth``iday`拼成日期 并结合`resolution`生成新字段`持续时间`
- `provstate` `city` `corp1` `target1` `gname` 将`string`转为类别编号
  
  
  首先实现2

```python
# 替换城市
cities = set(df_target.city)
dict_city = dict(zip(cities, list(range(len(cities)))))
df_target.city = df_target.city.apply(lambda x : dict_city.get(x))
df_target.city = pd.to_numeric(df_target.city)
print(len(cities), df_target.city.max())

# 替换 provstate
provstates = set(df_target.provstate)
dict_provstates = dict(zip(provstates, list(range(len(provstates)))))
df_target.provstate = df_target.provstate.apply(lambda x : dict_provstates.get(x))
df_target.provstate = pd.to_numeric(df_target.provstate)
print(len(provstates), df_target.provstate.max())

# 替换 corp1
corp1s = set(df_target.corp1)
dict_corp1s = dict(zip(corp1s, list(range(len(corp1s)))))
df_target.corp1 = df_target.corp1.apply(lambda x : dict_corp1s.get(x))
df_target.corp1 = pd.to_numeric(df_target.corp1)
print(len(corp1s), df_target.corp1.max())

# 替换 target1
target1s = set(df_target.target1)
dict_target1s = dict(zip(target1s, list(range(len(target1s)))))
df_target.target1 = df_target.target1.apply(lambda x : dict_target1s.get(x))
df_target.target1 = pd.to_numeric(df_target.target1)
print(len(target1s), df_target.target1.max())

# 替换 gname
gnames = set(df_target.gname)
dict_gnames = dict(zip(gnames, list(range(len(gnames)))))
df_target.gname = df_target.gname.apply(lambda x : dict_gnames.get(x))
df_target.gname = pd.to_numeric(df_target.gname)
print(len(gnames), df_target.gname.max())
```

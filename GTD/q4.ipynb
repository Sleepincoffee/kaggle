```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
df = pd.read_csv('/mnt/d/Documents/数学建模国赛/C题目/data/GTD_v2.csv')
df = df.set_index('eventid')
```

```python
years = list(range(1998,2018))
nos = [i*100000000+1010001 for i in years]

x = []
y = []
y_success = []
y_nkill = []

for i in range(len(nos)-1):
#     print(nos[i])
#     print(len(df.loc[nos[i]:nos[i+1], :].index))
    y.append(len(df.loc[nos[i]:nos[i+1], :].index))
    y_success.append(len(df[df.success==1].loc[nos[i]:nos[i+1], :].index))
    y_nkill.append(len(df[df.nkill>50].loc[nos[i]:nos[i+1], :].index))
    x.append(i+1998)


y.append(len(df.loc[nos[-1]:, :].index))
y_success.append(len(df[df.success==1].loc[nos[-1]:, :].index))
x.append(2017)

print(x,y,y_success)
```

```python
fig = plt.figure()
plt.plot(x,y, label='All')
plt.plot(x,y_success, label='Successful attack')
plt.xticks(x)
plt.legend()
plt.ylabel('Numbers')
fig.autofmt_xdate(rotation = 45)
plt.show()

print(x,y,y_success)
```

# 十大组织活动比例趋势图

```python
# 各地区

years = list(range(1998,2018))
nos = [i*100000000+1010001 for i in years]

x = []
y1 = []
y2 = []
y3 = []
y4 = []
y5 = []
y6 = []

for i in range(len(nos)-1):
#     print(nos[i])
#     print(len(df.loc[nos[i]:nos[i+1], :].index))
    n = len(df.loc[nos[i]:nos[i+1], :].index)
    n1 = len(df[df.gname == 'Islamic'].loc[nos[i]:nos[i+1], :].index)/n
    n2 = len(df[df.gname == 'Taliban'].loc[nos[i]:nos[i+1], :].index)/n
    n3 = len(df[df.gname == 'Al-Shabaab'].loc[nos[i]:nos[i+1], :].index)/n
    n4 = len(df[df.gname == 'Boko'].loc[nos[i]:nos[i+1], :].index)/n
    n5 = len(df[df.gname == 'New'].loc[nos[i]:nos[i+1], :].index)/n
    n6 = len(df[df.gname == 'Houthi'].loc[nos[i]:nos[i+1], :].index)/n
    
    x.append(i+1998)
    y1.append(n1)
    y2.append(n2)
    y3.append(n3)
    y4.append(n4)
    y5.append(n5)
    y6.append(n6)

n = len(df.loc[nos[i]:nos[i+1], :].index)
n1 = len(df[df.gname == 'Islamic'].loc[nos[-1]:, :].index)/n
n2 = len(df[df.gname == 'Taliban'].loc[nos[-1]:, :].index)/n
n3 = len(df[df.gname == 'Al-Shabaab'].loc[nos[-1]:, :].index)/n
n4 = len(df[df.gname == 'Boko'].loc[nos[-1]:, :].index)/n
n5 = len(df[df.gname == 'New'].loc[nos[-1]:, :].index)/n
n6 = len(df[df.gname == 'Houthi'].loc[nos[-1]:, :].index)/n

y1.append(n1)
y2.append(n2)
y3.append(n3)
y4.append(n4)
y5.append(n5)
y6.append(n6)
x.append(2017)

fig = plt.figure() 
plt.plot(x, y1, label='Islamic')
plt.plot(x, y2, label='Taliban')
plt.plot(x, y3, label='Al-Shabaab')
plt.plot(x, y4, label='Boko')
plt.plot(x, y5, label='NPA')
plt.plot(x, y6, label='Houthi')
plt.xticks(x)
plt.legend()
plt.ylabel('Proportion')
fig.autofmt_xdate(rotation = 45)
plt.title('Trends in terrorist events throughout the year')
plt.show()
print(y1,y2,y3)
```
